```python
import pandas as pd

df = pd.read_excel('base_transportes.xlsx')
df
```

```python
dict = {
    'Gás': [], 
    'Meio de Transporte': [], 
    'Combustivel': [], 
    'Tipo de Transporte': [], 
    'Atividade': [], 
    'Estado': [],
    'Emissão Total': [],
    'Ano': []
}

cont = 0

for loop in range(24):
    for coluna in ['Gás', 'Meio de Transporte', 'Combustivel', 'Tipo de Transporte', 'Atividade', 'Estado', 2000+loop]:
        if cont == 6:
            for linha in df[coluna]:
                dict['Emissão Total'].append(linha)
                dict['Ano'].append(2000+loop)
            cont = 0
            continue
        for linha in df[coluna]:
            dict[coluna].append(linha)
        cont += 1
        
        
```

```python
df_novo = pd.DataFrame(dict)
df_novo
```
